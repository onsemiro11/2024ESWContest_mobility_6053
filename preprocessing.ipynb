```python
from tqdm import tqdm
import os

images_path = '/home/hyundo/2024ESWContest_mobility_TRAVI/road_mark/train/images'
labels_path = '/home/hyundo/2024ESWContest_mobility_TRAVI/road_mark/train/labels'

ori_class = ['Pedestrian_Bicycle', 'Pedestrian_Pedestrian', 'RoadMarkArrow_Else', 'RoadMarkArrow_Left', 'RoadMarkArrow_Right', 'RoadMarkArrow_Straight', 'RoadMarkArrow_StraightLeft', 'RoadMarkArrow_StraightRight', 'RoadMarkArrow_Uturn', 'RoadMark_Character', 'RoadMark_Crosswalk', 'RoadMark_Number', 'RoadMark_StopLine', 'TrafficLight_Arrow', 'TrafficLight_Green', 'TrafficLight_GreenArrow', 'TrafficLight_Red', 'TrafficLight_RedArrow', 'TrafficLight_Yellow', 'TrafficSign_Else', 'TrafficSign_Speed', 'Vehicle_Bus', 'Vehicle_Car', 'Vehicle_Motorcycle', 'Vehicle_Unknown']
rmv_class = ['Pedestrian_Bicycle', 'RoadMark_Character', 'RoadMark_Number',  'Vehicle_Motorcycle', 'Vehicle_Unknown','TrafficSign_Else', 'TrafficSign_Speed']
new_class = ['Pedestrian_Pedestrian', 'RoadMarkArrow_Else', 'RoadMarkArrow_Left', 'RoadMarkArrow_Right', 'RoadMarkArrow_Straight', 'RoadMarkArrow_StraightLeft', 'RoadMarkArrow_StraightRight', 'RoadMarkArrow_Uturn', 'RoadMark_Crosswalk', 'RoadMark_StopLine', 'TrafficLight_Arrow', 'TrafficLight_Green', 'TrafficLight_GreenArrow', 'TrafficLight_Red', 'TrafficLight_RedArrow', 'TrafficLight_Yellow', 'Vehicle_Bus', 'Vehicle_Car']


def rename_files(images_path, labels_path):
    #     
    image_files = sorted([f for f in os.listdir(images_path) if f.endswith('.jpg')])
    label_files = sorted([f for f in os.listdir(labels_path) if f.endswith('.txt')])

    #    (      )
    assert len(image_files) == len(label_files), "      ."

    #   
    for idx, (img_file, lbl_file) in enumerate(zip(image_files, label_files)):
        new_name = idx  #    

        #  
        img_src_path = os.path.join(images_path, img_file)
        lbl_src_path = os.path.join(labels_path, lbl_file)

        #  
        img_dst_path = os.path.join(images_path, f"{new_name}.jpg")
        lbl_dst_path = os.path.join(labels_path, f"{new_name}.txt")

        #   
        os.rename(img_src_path, img_dst_path)
        os.rename(lbl_src_path, lbl_dst_path)

        print(f"Renamed: {img_file} -> {new_name}.jpg, {lbl_file} -> {new_name}.txt")


def process_label_file(file_path):
    with open(file_path, 'r') as f:
        labels = f.readlines()

    filtered_labels = []
    for label in labels:
        parts = label.split()
        if ori_class[int(parts[0])] not in rmv_class:
            new_class_id = new_class.index(ori_class[int(parts[0])])
            parts[0] = str(new_class_id)
            filtered_labels.append(' '.join(parts)+'\n')

    with open(file_path, 'w') as f:
        f.writelines(filtered_labels)

    return len(labels), len(filtered_labels)



rename_files(images_path, labels_path)

print("\n **** class remove **** \n")

total_processed = 0
total_kept = 0

for filename in os.listdir(labels_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(labels_path, filename)
        processed, kept = process_label_file(file_path)
        total_processed += processed
        total_kept += kept
        print(f"File {filename}: Processed {processed} labels, kept {kept}")

print(f"Total: Processed {total_processed} labels, kept {total_kept}")
```

```python
import os
import cv2
import random
from glob import glob
import numpy as np

def augment_image(image):
    #   (Random Brightness 20%)
    alpha = 1 + random.uniform(-0.2, 0.2)
    bright_image = cv2.convertScaleAbs(image, alpha=alpha, beta=0)
    return bright_image

def get_rare_classes(labels_path):
    class_counts = Counter()
    for label_file in os.listdir(labels_path):
        if label_file.endswith('.txt'):
            with open(os.path.join(labels_path, label_file), 'r') as f:
                labels = f.readlines()
                for label in labels:
                    class_id = int(label.split()[0])
                    class_counts[class_id] += 1
    
    #        
    avg_count = sum(class_counts.values()) / len(class_counts)
    rare_classes = set(class_id for class_id, count in class_counts.items() if count < avg_count)
    return rare_classes

#   
images_path = '/home/hyundo/2024ESWContest_mobility_TRAVI/road_mark/valid/images'
labels_path = '/home/hyundo/2024ESWContest_mobility_TRAVI/road_mark/valid/labels'

#    
rare_classes = get_rare_classes(labels_path)

#    
image_files = glob(os.path.join(images_path, '*.jpg'))

#  : 3 
augmentation_factor = 3

print(f"Starting selective augmentation for rare classes: {rare_classes}")

#     
for img_file in image_files:
    base_name = os.path.basename(img_file).replace('.jpg', '')
    label_file = os.path.join(labels_path, f"{base_name}.txt")
    
    #       
    if os.path.exists(label_file):
        with open(label_file, 'r') as f:
            labels = f.readlines()
        if any(int(label.split()[0]) in rare_classes for label in labels):
            image = cv2.imread(img_file)
            
            for i in range(augmentation_factor - 1):
                #  
                augmented_image = augment_image(image)

                #   
                new_img_name = f"aug_{base_name}_{i}.jpg"
                new_img_path = os.path.join(images_path, new_img_name)
                
                new_label_name = f"aug_{base_name}_{i}.txt"
                new_label_path = os.path.join(labels_path, new_label_name)

                #   
                cv2.imwrite(new_img_path, augmented_image)
                
                #   
                with open(new_label_path, 'w') as f:
                    f.writelines(labels)
                
                print(f"Saved: {new_img_path} and {new_label_path}")

print("Selective data augmentation complete.")
```
